#### Parmis les fonctions j'output quelques statistques descriptives de la base telles qu'une matrice de corr√©lation et descriptions 

```python
dp = DataPipeline(data_filename='assurance_vie.csv')
dp.get_data()
dp.processing()
dp.viz_corr()
dp.dumies_modalite()
dp.split_train()
dp.X_train
```

#### J'entraine, je pr√©dis et j'√©value trois mod√®les => Regression Logistic, Random Forest, Decision tree afin de comparer leurs performances 

```python
lr=ModelPipeline()
lr.train(dp.X_train, dp.y_train)
lr_pred=lr.predict(dp.X_test)
lr_pred_soft = lr.predict_proba(dp.X_test)
lr_score=lr.evaluate(dp.y_test,lr_pred)
lr_fpr_model, lr_tpr_model, lr_auc=lr.roc(dp.y_test, lr_pred_soft)
```

```python
rf=ModelPipeline(model=RandomForestClassifier(min_samples_split=50, n_estimators=200) )
rf.train(dp.X_train, dp.y_train)
rf_pred=rf.predict(dp.X_test)
rf_pred_soft = rf.predict_proba(dp.X_test)
rf_score=rf.evaluate(dp.y_test,rf_pred)
rf_fpr_model, rf_tpr_model, rf_auc=rf.roc(dp.y_test, rf_pred_soft)
```

```python
dt=ModelPipeline(model=DecisionTreeClassifier(max_depth=4, random_state=42) )
dt.train(dp.X_train, dp.y_train)
dt_pred=rf.predict(dp.X_test)
dt_pred_soft = dt.predict_proba(dp.X_test)
dt_score=rf.evaluate(dp.y_test,rf_pred)
dt_fpr_model, dt_tpr_model, dt_auc=rf.roc(dp.y_test, dt_pred_soft)
```

#### Je trace la courbe ROC qui est un indicateur de pr√©diction des vrais valeurs (true positif), l'air sous la courbe nous donne une indication sur la capacit√© des mod√®les √† identifier les positifs et n√©gatifs

```python
# Tracage des courbes ROC pour chaque mod√®le
plt.figure(figsize=(8, 6))

plt.plot(lr_fpr_model, lr_tpr_model, color='darkorange', lw=2, label='Regression logit (AUC = %0.2f)' % lr_auc)
plt.plot(rf_fpr_model, rf_tpr_model, color='blue', lw=2, label='Random forest (AUC = %0.2f)' % rf_auc)
plt.plot(dt_fpr_model, dt_tpr_model, color='green', lw=2, label='Decision tree (AUC = %0.2f)' % dt_auc)


plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
